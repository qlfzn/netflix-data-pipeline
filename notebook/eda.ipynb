```python
import pandas as pd
import os

source_path = "../data/"

dataframes = {}

for file in os.listdir(source_path):
    name = file.split(".")[0]
    ext = file.split(".")[1]

    if ext != "csv":
        continue

    df= pd.read_csv(source_path+file)
    dataframes[name] = df

    print(f"\nDataframe {name} successfully created")
    print(f"Created with {df.shape[0]} rows & {df.shape[1]} columns")

    
```

```python
# check columns
for name, df in dataframes.items():
    print(f"\n--{name}--\n", df.dtypes)
```
